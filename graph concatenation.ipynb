```python
%matplotlib inline


def plt_graph(sent):
    g = GraphBuilder()
    g.gen_data_one(sent)
    parse_tree = tree_gen.get_tree(sent)
    graph = g.gen_graph_one(parse_tree)
    g.plot_graph(graph)
    del g
    return
```

```python
plt_graph("I am a boy")
```

```python
plt_graph("She is a good girl")
```

```python
%matplotlib inline

G = GraphBuilder()
sents = ["I am a boy", "She is a good girl"]
G.gen_data(sents)
giant_graph = G.gen_giant_graph()
G.plot_graph(giant_graph)
```
